**Interpolação linear**

```python
dataset_co2_FPAIS.replace("DESCONHECIDO", np.nan, inplace=True)
dataset_co2_FBLOCO.replace("DESCONHECIDO", np.nan, inplace=True)
dataset_co2_FCONTINENTE.replace("DESCONHECIDO", np.nan, inplace=True)

dataset_co2_FBLOCO.to_excel('co2_emissions_global_groups.xlsx', index=False)
dataset_co2_FPAIS.to_excel('co2_emissions_global_countries.xlsx', index=False)
dataset_co2_FCONTINENTE.to_excel('co2_emissions_global_continents.xlsx', index=False)
```

```python
numeric_columns_FPAIS = dataset_co2_FPAIS.columns.difference(['país', 'ano'])
numeric_columns_FBLOCO = dataset_co2_FBLOCO.columns.difference(['país', 'ano'])
numeric_columns_FCONTINENTE = dataset_co2_FCONTINENTE.columns.difference(['país', 'ano'])

dataset_co2_FPAIS[numeric_columns_FPAIS] = dataset_co2_FPAIS[numeric_columns_FPAIS].apply(pd.to_numeric, errors='coerce')
dataset_co2_FBLOCO[numeric_columns_FBLOCO] = dataset_co2_FBLOCO[numeric_columns_FBLOCO].apply(pd.to_numeric, errors='coerce')
dataset_co2_FCONTINENTE[numeric_columns_FCONTINENTE] = dataset_co2_FCONTINENTE[numeric_columns_FCONTINENTE].apply(pd.to_numeric, errors='coerce')
```

```python
def interpolate_group(group):
    return group.interpolate(method='linear', limit_direction='both')

dataset_co2_FPAIS = dataset_co2_FPAIS.groupby('país').apply(interpolate_group)
dataset_co2_FBLOCO = dataset_co2_FBLOCO.groupby('país').apply(interpolate_group)
dataset_co2_FCONTINENTE = dataset_co2_FCONTINENTE.groupby('país').apply(interpolate_group)

dataset_co2_FBLOCO.to_excel('co2_emissions_global_groups.xlsx', index=False)
dataset_co2_FPAIS.to_excel('co2_emissions_global_countries.xlsx', index=False)
dataset_co2_FCONTINENTE.to_excel('co2_emissions_global_continents.xlsx', index=False)
```
